```python
#Deleting unnecessary columns
```

```python
df1.drop(labels='Photo',axis=1,inplace=True)
df1.drop(labels='Loaned From',axis=1,inplace=True)
df1.drop(labels='Body Type',axis=1,inplace=True)
df1.drop(labels='Jersey Number',axis=1,inplace=True)
df1.drop(labels='Club Logo',axis=1,inplace=True)
df1.drop(labels='Flag',axis=1,inplace=True)
```

```python
df1.describe()
```

```python
#Treating the Value and Wage Column 
```

```python
df1['Value'].isnull().sum()
```

```python
df1['Wage'].isnull().sum()
```

```python
df1['Release Clause'].isnull().sum()
```

```python
df1['Release Clause'].mode()
```

```python
df1['Release Clause'].fillna("€1.1M",inplace=True)#Imputing the mode for Release Clause
```

```python
def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
df1['Value'] = df1['Value'].apply(lambda x: extract_value_from(x))
```

```python
df1['Wage'] = df1['Wage'].apply(lambda x: extract_value_from(x))
```

```python
df1['Release Clause']=df1['Release Clause'].apply(lambda x: extract_value_from(x))
```

```python
#Treating the Joined column
```

```python
temp=[]
```

```python
for i in range(len(df1['Joined'].values)):
    if isinstance(df1['Joined'].values[i], float):
        temp.append(np.nan)
    else:
        temp.append(df1['Joined'].values[i].split(', ')[1])
```

```python
temp2=list(map(float,temp))
```

```python
df1.drop(labels='Joined',inplace=True,axis=1) #Dropping the oroginal column
```

```python
df1["Joined"]=temp2 #imputing the treated list as the new data for the joined column
```

```python
df1.Joined.fillna(int(np.mean(df1['Joined'])),inplace=True) #filling nans with mean and converting it in int
```

```python
#Treating Height Column
```

```python
df1['Height'].isnull().sum()
```

```python
temph=[]
```

```python
for i in range(len(df1['Height'].values)):
    if isinstance(df1['Height'].values[i], float):
        temph.append(np.nan)
    else:
        temph.append(df1['Height'].values[i].replace("'","."))
```

```python
temph2=list(map(float,temph))
len(temph2)
```

```python
df1.drop(labels='Height',inplace=True,axis=1)#Dropping oroginal Height Column
```

```python
df1['Height']=temph2#imputing the treated list as the new data for the joined column
```

```python
df1.Height.fillna(np.mean(df1['Height']),inplace=True) #filling nans with the mean 
```

```python
#Treating Contract Valid Until column 
```

```python
df1['Contract Valid Until']= pd.to_datetime(df1['Contract Valid Until'], errors='coerce')
```

```python
#Treating Weight column
```

```python
df1['Weight'].isnull().sum()
```

```python
tempw=[]
```

```python
for i in range(len(df1['Weight'].values)):
    if isinstance(df1['Weight'].values[i], float):
        tempw.append(np.nan)
    else:
        tempw.append(df1['Weight'].values[i].replace("lbs",""))
```

```python
tempw2=list(map(float,tempw))
```

```python
df1.drop(labels='Weight',inplace=True,axis=1)
```

```python
df1["Weight"]=tempw2
```

```python
df1["Weight"].isnull().sum()
```

```python
df1.fillna(np.mean(df1["Weight"]),inplace=True)
```

```python
#Generate tables containing the top 5 players by Overall rating for each unique position.
```

```python
top_5 = {}
for i in df1['Position'].unique():
    try:
        top_5[i]=list(df1.sort_values(by=['Overall','Position'], ascending=[False, True]).groupby(df1['Position']).get_group(i)[:5]['Name'].values)
    except KeyError:
        pass
```

```python
len(top_5.keys())
```

```python
len(df1['Position'].unique())
```

```python
df_top_5 = pd.DataFrame(top_5)
df_top_5
```

```python
df_top_5.duplicated()
```

```python
#No players are appearing in morethan one table
```

```python
#What is the average wage one can expect to pay for the top 5 in every position?
```

```python
sal= {}
for i in df_top_5.columns:
    h= []
    for j in df_top_5[i]:
        try:
            h.append(*df1.loc[df1['Name'] == j]['Wage'].values)
        except TypeError:
            pass
        sal[i]= sum(h)
```
